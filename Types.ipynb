### Неизменяемые типы данных

К неизменяемым типам данных относят: целые числа (int), вещественные числа (float), комплексные числа (complex), логические переменные (bool), 
кортежи (tuple), неизменяемые множества (frozen set). 
При задании переменной создается объект с полем данных нужного типа, с уникальным идентификатором, 
затем созданная переменная начинается ссылаться на этот объект. 
Неизменяемость типа данных не означает, что переменной нельзя присвоить новое значение,
это означает, что при присвоении нового значения будет создан новый объект, и данная переменная начнет ссылаться на него
(передача объекта по значению).

```python
# срезы (slicing)
# имя[начало:конец:шаг] - любое значение может отсутствовать
l3=list(range(10)) #range() - функция генерации натуральных чисел из встроенной библиотеки 
print(l3)
print(l3[2:5]) #со втрого (третьего) до пятого (четвертого)
print(l3[:6])  #с начала списка, шаг==1
print(l3[2:])  #до конца списка, шаг==1
print(l3[::3]) #с начала до конца списка, шаг==1
print(l3[::-1]) # c конца до начала списка, шаг==1
```

## Циклы

```python
#Цикл while (аналогично такому циклу в С)
k=10
while k>0:
    print('*')
    k-=1
```

```python
print("Ввод текста. Вывод  прописными бкуквами. Для окончания и вводе '*'")
while True:
    inp=input()
    if inp=='*':
        break
    print(inp.upper())
# break - прерывание цикла
# continue - пропуск итерации
```

```python
#цикл for - работа с итерабельными объектами
count=['one','two','three','four','five']
for i in count:
    print(i)
print ('\n\nВ обратном порядке:')
for i in count[::-1]:
    print(i)
```

```python
str="абракадабра"
for i in str:
    print(i)
```

```python
# итерации в for  по нескольким переменным (функция zip())
count_en=['one','two','three','four','five']
count_rus=['раз','два','три','четыре','пять',]
cc=list(range(1,6,1))
for i,j,k in zip(cc, count_rus, count_en):
    print(i, ' -> ', j, '->', k )
print("Вышел зайчик погулять...".upper())    
```
