```python
# for -    
count=['one','two','three','four','five']
for i in count:
    print(i)
print ('\n\n  :')
for i in count[::-1]:
    print(i)
```

```python
#   for     ( zip())
count_en=['one','two','three','four','five']
count_rus=['','','','','',]
cc=list(range(1,6,1))
for i,j,k in zip(cc, count_rus, count_en):
    print(i, ' -> ', j, '->', k )
print("  ...".upper())    
```
